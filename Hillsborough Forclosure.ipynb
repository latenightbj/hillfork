```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="whambamalam")
```

```python
import time
```

```python
z =  pd.read_csv('data/nocondounit.csv')

# for index, row in df.iterrows():
#     print(row['c1'], row['c2'])
for row in z.iterrows():
    time.sleep(1)
    location = geolocator.geocode(row[1]["Address"] + " " + str(row[1]["Zip"]))
    try:
        print((location.latitude, location.longitude))
    except:
        pass
# print(geolocator.geocode("409 13TH ST NE, 33570"))
```

```python
import folium

map1 = folium.Map(
    location=[27.964157,-82.452606],
    tiles='cartodbpositron',
    zoom_start=12,
)
```

```python
for row in z.iterrows():
#   time.sleep(1)
    location1 = geolocator.geocode(row[1]["Address"] + " " + str(row[1]["Zip"]))
    try:        
        folium.CircleMarker(location=[location1.latitude, location1.longitude]).add_to(map1)
    except:
        print('unfound address')
        pass
```
